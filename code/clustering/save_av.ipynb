```python
import imageio.v2 as imageio
import os

#methods = ["redcap", "clustergeo", "skater"]
methods = ["redcap"]
seasons = ["2023-24", "2024-25", "2025-26"]

base = "/work2/09967/dongahkim0223/frontera/Spatial_clustering"

for method in methods:
    for season in seasons:
        files = [
            f"{base}/figures/cluster_combine/county_{method}_exclude_{season}_k{i}.png"
            for i in range(5, 46, 2)
        ]

        images = []
        missing = False

        for f in files:
            if not os.path.exists(f):
                print("MISSING:", f)
                missing = True
                break
            print("READ:", f)
            images.append(imageio.imread(f))

        if missing:
            print(f"Skipping {method}-{season}")
            continue

        outfile = f"{base}/figures/county_{method}_{season}_animation.gif"
        imageio.mimsave(outfile, images, duration=1.0)

        print("SAVED:", outfile)
```
